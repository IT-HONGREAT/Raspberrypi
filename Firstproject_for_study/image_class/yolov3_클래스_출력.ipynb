```python
data = pd.read_csv('/content/음식이미지 영양정보 402.csv')
data
```

```python
def classifi(input_class, input_weight):
  for i in range(len(data)):
    if data.iloc[i,6] == input_class:
      nutrition = data.iloc[i]
  div = input_weight/nutrition.iloc[12]
  print('입력받은 음식 무게는',input_weight,'g입니다.')
  print('1회 제공량의 약',round(div,2),'배 입니다.')
  print('식품명:',nutrition.iloc[6])
  print('칼로리:',round(nutrition.iloc[16]*div,2),'kcal')
  print('탄수화물',round(nutrition.iloc[23]*div,2),'g')
  print('단백질', round(nutrition.iloc[20]*div,2),'g')
  print('지방', round(nutrition.iloc[21]*div,2),'g')
```

# 좌표 값

```python
coordi = open('/content/result.txt').read().splitlines()
print(coordi)
```

```python
#정규 표현식
import re
p = re.compile('^[가-힣a-zA-A0-9]*:')
```

```python
result_list = []
class_names = []
for text in coordi:
  result = p.findall(text)
  if len(result) != 0:
    result_list.append(result)
result_list = sum(result_list, [])

for i in range(len(result_list)):
  print(result_list[i][:-1])
  input_class= result_list[i][:-1]
  class_names.append(input_class)
print(class_names)
```

```python
text_list = []
for text in coordi:
  result = p.findall(text)
  if len(result) != 0:
    print(text)
    text_list.append(text)

print(text_list)
```

https://codechacha.com/ko/python-extract-integers-from-string/ 문자열에서 숫자 반환

```python
coordinate = []
class_percent = []
for n in range(len(text_list)):
  numbers = re.findall(r'\d+', text_list[n])
  coordinate.append(numbers[1:5])
  class_percent.append(numbers[0])
print("좌표값",coordinate) 
print("정확도",class_percent)
print("클래스 명",class_names)


print("")
for i in range(len(class_names)):
  print(f'{class_names[i]}의 정확도는 {class_percent[i]}% 이며 좌표값은 {coordinate[i]}입니다.')
```
